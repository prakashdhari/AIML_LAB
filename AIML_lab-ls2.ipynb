```python
import matplotlib.pyplot as plt
```

```python
y_Apples=[0.895,0.91,0.919,0.926,0.929,0.931]
```

```python
plt.plot(y_Apples)
```

```python
years=[2010,2011,2012,2013,2014,2015]
plt.plot(years,y_Apples)
```

```python
plt.plot(years,y_Apples)
plt.xlabel('Years')
plt.ylabel('Apples yield (tons per hectare)')
plt.title('Apples in the UK')
```

```python
years=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937]
oranges=[0.962,0.941,0.930,0.923,0.918,0.907,0.905,0.901]
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.title('Apples vs. oranges in the UK')
plt.legend(['Apples','Oranges'])
```

```python
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.title('Apples vs. oranges in the UK')
plt.legend(['Apples','Oranges'])
```

```python
plt.plot(years,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='navy')
plt.plot(years,oranges,marker='o',c='r',ls='--',lw=3,ms=10,alpha=0.5)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.title('Apples vs. oranges in the UK')
plt.legend(['Apples','Oranges'])
```

```python
plt.plot(years,apples,'s-b')
plt.plot(years,oranges,'o--r')
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.title('Apples vs. oranges in the UK')
plt.legend(['Apples','Oranges'])
```

```python
plt.plot(years,oranges,'or')
plt.title('Oranges in the UK')
```

```python
plt.figure(figsize=(4,4))
plt.plot(years,oranges,'or')
plt.title('Oranges in the UK')
```

```python
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.bar(x,y)
plt.title('Bar chart')
plt.legend(['Bar'])
plt.show()
```

```python
plt.scatter(x,y)
plt.title('Scatter chart')
plt.legend("A")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x=[1,2,3,4]
e=(0.1,0.2,0.3,0.2)
plt.pie(x,explode=e)
plt.title('Pie chart')
plt.show()
```
